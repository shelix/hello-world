```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint as rand
%matplotlib inline
```

First of all, how random is randint?

```python
x = np.random.randint(-100,100,size=500)
dist=plt.hist(x)
```

OK, so it looks uniform. Now let's generate a random process using a simple autoregressive model:

```python
y = np.zeros(50)
for j in range(1,50-1):
    y[j+1]=y[j]+rand(-10,10)
plt.plot(y)
```

The proces slowly drifts around. Now generate multiple random processes. Where do all these end up going?

```python
Ny = 20
Tmax = 100
emax = 10
y = np.zeros((Tmax,Ny))
for i in range(0,Ny-1):
    for j in range(0,Tmax-1):
        y[j+1,i]=y[j,i]+rand(-emax,emax)
    plt.plot(y)
```

At a later point in time, the value of each random variable is a sum of randomly generated numbers. 
So by the Central Limit Theorum, the distribution of all the values at a later point in time should approach
the normal distribution. Let's try it, but with larger numbers:

```python
Ny = 100
Tmax = 500
emax = 10
y = np.zeros((Tmax,Ny))
for i in range(0,Ny-1):
    for j in range(0,Tmax-1):
        y[j+1,i]=y[j,i]+rand(-emax,emax)
plt.show(plt.hist(y[100]))
```

Cool, that looks kind of normal!
